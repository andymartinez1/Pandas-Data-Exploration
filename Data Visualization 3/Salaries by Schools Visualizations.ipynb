```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv(
    r"C:\Python\Pandas\Pandas Data Exploration 3\Data Visualization 3\salaries-by-region.csv"
)
pd.set_option("display.max.columns", None)
```

```python
df["Starting Median Salary"] = pd.to_numeric(df["Starting Median Salary"])
df["Mid-Career Median Salary"] = pd.to_numeric(df["Mid-Career Median Salary"])
df["Mid-Career 10th Percentile Salary"] = pd.to_numeric(
    df["Mid-Career 10th Percentile Salary"]
)
df["Mid-Career 25th Percentile Salary"] = pd.to_numeric(
    df["Mid-Career 25th Percentile Salary"]
)
df["Mid-Career 75th Percentile Salary"] = pd.to_numeric(
    df["Mid-Career 75th Percentile Salary"]
)
df["Mid-Career 90th Percentile Salary"] = pd.to_numeric(
    df["Mid-Career 90th Percentile Salary"]
)
```

```python
# Check column data types after numeric conversion
print(df.dtypes)
```

```python
# Used print() here because it was not displaying the first few rows
print(df.head())
```

```python
# Converting Nan to 0s
df.fillna(0)
```

```python
# Using different visualization techniques
starting_median_salary = df.nlargest(10, "Starting Median Salary")
starting_median_salary.plot(x="School Name", y="Starting Median Salary", kind="scatter", rot=90)

mid_career_salary = df.nlargest(10, "Mid-Career Median Salary")
mid_career_salary.plot(x="School Name", y="Mid-Career Median Salary", kind="barh")
mid_career_salary.plot(x="School Name", y="Mid-Career Median Salary", kind="box")
```
