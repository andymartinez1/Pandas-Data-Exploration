```python
# -*- coding: utf-8 -*-
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Setting the path to where the file is located
path = r"2021_rws.xlsx"
```

```python
# Reading the .csv file
remote_worker_questions = pd.read_excel(path)
```

```python
# Selecting what I find to be the most relevant data from the database
useful_remote_questions = remote_worker_questions.iloc[
    :, [4, 5, 6, 11, 12, 32, 47, 93, 94, 95, 107]
]
print(useful_remote_questions)
```

```python
# Relabeling the columns to shorter, more accurate headers
useful_remote_questions.columns = [
    "Industry",
    "Occupation",
    "Employee_Size",
    "Time_Worked_Remotely",
    "Remote_Preference",
    "Positivity",
    "Pay_Cut",
    "Better_Working",
    "More_Active",
    "In_Office",
    "Productivity",
]
```

```python
# Plotting and labeling the graph
print(useful_remote_questions["Better_Working"].value_counts())
plt.plot(useful_remote_questions["Better_Working"].value_counts())
plt.xlabel("x - axis")
plt.xticks(rotation=90)
plt.ylabel("Voter Counts")
plt.show()
```
